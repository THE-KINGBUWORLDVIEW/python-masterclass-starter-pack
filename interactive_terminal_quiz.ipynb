# Project 1: Interactive Python Quiz 🚀

Welcome to your first Python project! In this notebook, we will build a command-line trivia game.

**What you will learn today:**
* **Variables:** Storing information.
* **Data Types:** Working with Lists and Dictionaries.
* **Loops:** Automating repetitive tasks.
* **Conditionals:** Making decisions with code (if/else).

```python
# 1. Setting up our trivia data
# We use a LIST (denoted by []) to hold multiple items.
# Inside the list, we use DICTIONARIES (denoted by {}) to store key-value pairs.

quiz_questions = [
    {
        "question": "What is the output of print(2 ** 3)?",
        "options": ["A) 6", "B) 8", "C) 9", "D) Error"],
        "answer": "b"
    },
    {
        "question": "Which data type is used to store True or False values?",
        "options": ["A) String", "B) Integer", "C) Boolean", "D) Float"],
        "answer": "c"
    },
    {
        "question": "What symbol is used to start a comment in Python?",
        "options": ["A) //", "B) <!-- -->", "C) /*", "D) #"],
        "answer": "d"
    },
    {
        "question": "Which built-in function returns the length of a list?",
        "options": ["A) length()", "B) len()", "C) count()", "D) size()"],
        "answer": "b"
    },
    {
        "question": "What is the correct file extension for Python files?",
        "options": ["A) .python", "B) .pyt", "C) .py", "D) .pt"],
        "answer": "c"
    },
    {
        "question": "How do you create a variable named 'x' with the numeric value 5?",
        "options": ["A) int x = 5", "B) x = 5", "C) x : 5", "D) let x = 5"],
        "answer": "b"
    },
    {
        "question": "Which of these collections is ordered and allows duplicate members?",
        "options": ["A) List", "B) Set", "C) Dictionary", "D) None of the above"],
        "answer": "a"
    },
    {
        "question": "Which keyword is used to create a function in Python?",
        "options": ["A) function", "B) create", "C) def", "D) make"],
        "answer": "c"
    },
    {
        "question": "What is the output of print(10 % 3)?",
        "options": ["A) 3", "B) 1", "C) 3.33", "D) 0"],
        "answer": "b"
    },
    {
        "question": "Which string method removes whitespace from the beginning and end?",
        "options": ["A) clean()", "B) trim()", "C) remove()", "D) strip()"],
        "answer": "d"
    }
]
```

```python
# 2. Initializing the game state

# Create a variable to keep track of the player's score
score = 0

# Welcome the player
print("Welcome to the KingbuWorldView Python Masterclass Quiz!")
print("Let's see how much you know...")
print("-" * 40) # Prints a neat dividing line
```

```python
# 3. The Main Game Loop
# We use a FOR loop to go through each question one by one.
# 'enumerate' helps us keep track of the question number.

for index, q in enumerate(quiz_questions):
    # Print the question number and the question itself
    print(f"\nQuestion {index + 1}: {q['question']}")

    # Print all the options associated with the current question
    for option in q['options']:
        print(option)

    # Get the user's answer.
    # .strip() removes accidental spaces, and .lower() makes it case-insensitive.
    user_answer = input("Your answer (A, B, C, or D): ").strip().lower()

    # Conditional check: Is the answer correct?
    if user_answer == q['answer']:
        print("Correct!")
        score += 1  # Add 1 to the score
    else:
        # If wrong, show the correct answer. .upper() capitalizes it for display.
        print(f"Incorrect. The correct answer was {q['answer'].upper()}.")

    print("-" * 40) # Divider for the next question
```

```python
# 4. Final Score and Conclusion

print("\n--- Quiz Complete! ---")
print(f"Your final score is: {score} out of {len(quiz_questions)}")

# Calculate the percentage
percentage = (score / len(quiz_questions)) * 100

# Provide conditional feedback based on how well they did
if percentage >= 80:
    print("Outstanding! You're a Python master! 🏆")
elif percentage >= 70:
    print("Great job! You have a solid grasp of the basics. 🌟")
else:
    print("Good effort! Keep practicing and you'll get there! 💪")
```
